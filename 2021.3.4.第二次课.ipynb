```python
# import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

x = np.linspace(0.05,10.1000)  #linspace能够精确控制终止值终值，而arange能够更直接地控制序列中值之间的增量。
y = np.sin(x)
plt.plot(x,y,lw=2,c='b',label='sin fig')#plt.plot(x, y, format_string, **kwargs)
# https://blog.csdn.net/weixin_44010756/article/details/109326364
plt.legend()
# 函数功能：给图形加上图例。
# matplotlib.pyplot.legend(*args, **kwargs)
# loc：图例在图中的地理位置
# 如 plt.legend（loc=‘upper left’）图例在左上角
# https://blog.csdn.net/Mr_blueD/article/details/88210961
plt.annotate('sin curve',xy=(np.pi/2,1.0),xytext=((np.pi/2+2.0,0.5)),weight='bold',
            color='r',arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=0.2',color='g'))
# s 为注释文本内容
# xy 为被注释的坐标点
# xytext 为注释文字的坐标位置
# xycoords 参数如下......
# weight 设置字体线型
# color 设置字体颜色
# arrowprops #箭头参数,参数类型为字典dict
# bbox给标题增加外框 ，常用参数如下.....
# https://blog.csdn.net/qq_36387683/article/details/101377416
```

```python
x = np.arange(0,10,1)
y = 2*x
for a,b in zip(x,y):
    plt.text(a,b+0.5,(a,b),fontsize=12,ha='center',va='bottom')#plt.text()函数用于设置文字说明。
#     https://blog.csdn.net/TeFuirnever/article/details/88947248
plt.plot(x,y,'bo-')
```
